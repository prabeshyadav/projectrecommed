```python
import pandas as pd
from sklearn.metrics import pairwise_distances
```

```python
frame = pd.read_csv('quantified_job_data.csv')
frame.set_index('Jobid', inplace=True)
frame.head()
```

```python
cosine = 1 - pairwise_distances(frame, metric='cosine')
new_frame = pd.DataFrame(cosine)
new_frame.head()
```

```python
first_user = new_frame.iloc[[0]]
similar_users = first_user.stack().sort_values(ascending=False)[0:10:1]
similar_users
```
