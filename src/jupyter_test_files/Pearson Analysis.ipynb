```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from copy import deepcopy
```

```python
initial_data = pd.read_csv('dataset.csv')
initial_data.head()
```

```python
data = deepcopy(initial_data)
data.head()
```

```python
company = data['Company']
company.head()
```

```python
data.drop('Company', axis=1, inplace=True)
data.set_index('Id', inplace=True)
data.head()
```

```python
pearson_sim = 1-pairwise_distances(data, metric="correlation")
pearson_results = pd.DataFrame(pearson_sim)
pearson_results.head()
```

```python
cosine_sim = 1 - pairwise_distances(data, metric="cosine")
cosine_results = pd.DataFrame(cosine_sim)
cosine_results.head()
```
